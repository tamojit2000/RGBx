```python
def makeColorImg(name='test.jpg',x=10,y=10,colour=tuple([0,0,0])):
    R,G,B=colour
    content=y*x*3
    IMG=np.zeros(content,dtype=int)
    i=0
    while(i<content):
        IMG[i]=np.random.randint(0,255)
        i+=1
        IMG[i]=np.random.randint(0,255)
        i+=1
        IMG[i]=np.random.randint(0,255)
        i+=1
    IMG=IMG.reshape(y,x,3)
    cv2.imwrite(name,IMG)
    #cv2.imshow('image',IMG)
    #cv2.waitKey()
    return IMG
    
    
    
```
