```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('amsterdam-1920x1080-channel-4k-18504.jpg')
```

```python
plt.imshow(img)
```

```python
y,x,a=img.shape
```

```python
for i in range(y):
    for j in range(x):
        BGR=img[i][j]
        BGR=[255-BGR[0],255-BGR[1],255]
        img[i][j]=BGR
```

```python
def HEXtoRGB(HEX):
    h=HEX.lstrip('#')
    return [int(h[i:i+2],16) for i in range(0,6,2)]
```
